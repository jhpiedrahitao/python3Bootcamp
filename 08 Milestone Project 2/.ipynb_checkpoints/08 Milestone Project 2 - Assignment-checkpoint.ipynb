```python
import random 
tipos = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
rango = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':1}

class Card:
    def __init__(self, tipo, rango,hidden=False):
        self.hidden=hidden
        self.tipo=tipo
        self.rango=rango
        self.valor=values[rango]
    def __str__(self):
        if self.hidden:
            return "hidden card"
        else:
            return f'{self.rango} of {self.tipo}'
        
class Deck():
    def __init__(self):
        
        self.all_cards = []
        
        for tipo in tipos:
            for rangoi in rango:
                self.all_cards.append(Card(tipo,rangoi))
    
    def __str__(self):
        cardlist=''
        for card in self.all_cards:
            cardlist+=str(card)+'\n'
        return cardlist
   
    def shuffle(self):
        random.shuffle(self.all_cards)
    
    def deal_one(self):
        return self.all_cards.pop()
    
class Player:
   
    def __init__(self,name):
        self.name=name
        self.all_cards=[]
        
    def __str__(self):
        cardlist=''
        for card in self.all_cards:
            cardlist+=str(card)+'\n'
        return f'player {self.name} has {len(self.all_cards)} cards \n'+cardlist
    
    def add_cards(self,newcards,hidden=False):
        newcards.hidden=hidden
        self.all_cards.append(newcards)
    def total(self):
        suma=[0]
        az=False
        for card in self.all_cards:
            if card.valor==1:
                az=True;
                suma[0]+= card.valor
                suma.append(1)
            else:
                suma[0]+= card.valor
        return suma
    
class Dealer(Player):
    def __init__(self):
        self.name='dealer'
        self.all_cards=[]
    def show(self):
        for card in self.all_cards:
            card.hidden=False
```

```python
#init setup
player_one=Player('one')
dealer=Dealer()
newdeck=Deck()
newdeck.shuffle()

player_one.add_cards(newdeck.deal_one())
dealer.add_cards(newdeck.deal_one(),hidden=False)
player_one.add_cards(newdeck.deal_one())
dealer.add_cards(newdeck.deal_one(),hidden=True)

game_on=True
pasa=False
# loop
while game_on:

    print(player_one)
    print(dealer)

    if player_one.total()[0]>21:
        dealer.show()
        print(player_one)
        print(dealer)
        print('jugador uno perdio')
        game_on=False
        break
    
    if player_one.total()[0]==21:
        dealer.show()
        print(player_one)
        print(dealer)
        print('jugador uno gano')
        game_on=False
        break
        
    if pasa==True:
        dealer.show()
        print(player_one)
        print(dealer)
        #print(len(player_one.total()))
        if len(player_one.total())==1:
            if player_one.total()[0]>dealer.total()[0]:
                print('jugador uno gana')
            elif player_one.total()[0]==dealer.total()[0]:
                print('empate')
            elif player_one.total()[0]<dealer.total()[0]:
                print('jugador uno pierde')
            game_on=False
            break
        else:
            print('azes en el juego')
            game_on=False
            break

    while True:
        other=True
        sn=input('quieres otra carta? s o n ')
        if sn.lower()=='n':
            other=False
            break
        elif sn.lower()=='s':
            other=True
            break
        else: 
            print('invalido')
            continue
    if other==True:
        player_one.add_cards(newdeck.deal_one()) 
        if dealer.total()[0]<17:
            dealer.add_cards(newdeck.deal_one(),hidden=True)
        else:
            continue
    elif other==False:
        pasa=True
        dealer.show()
        continue
```
